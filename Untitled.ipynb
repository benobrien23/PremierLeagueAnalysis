```python
# ==========================================================
# 0. Crest Dictionary (YOUR FILE NAMES)
# ==========================================================

crest_paths = {
    "Arsenal": "Arsenal-FC-logo-480x480.png",
    "Aston Villa": "Aston-Villa-FC-logo-480x480.png",
    "Bournemouth": "AFC-Bournemouth-logo-480x480.png",
    "Brentford": "Brentford-FC-logo-480x480.png",
    "Brighton": "Brighton-Hove-Albion-logo-480x480.png",
    "Chelsea": "Chelsea-FC-logo-480x480.png",
    "Cry. Palace": "Crystal-Palace-FC-logo-480x480.png",
    "Everton": "Everton-FC-logo-480x480.png",
    "Fulham": "Fulham-FC-logo-480x480.png",
    "Liverpool": "Liverpool-FC-logo-480x480.png",
    "Man. City": "Manchester-City-FC-logo-480x480.png",
    "Man. Utd": "Manchester-United-FC-logo-480x480.png",
    "Newcastle": "Newcastle-United-logo-480x480.png",
    "Nott Forest": "Nottingham-Forest-FC-logo-480x480.png",
    "Spurs": "Tottenham-Hotspur-logo-480x480.png",
    "West Ham": "West-Ham-United-FC-logo-480x480.png",
    "Wolves": "Wolverhampton-Wanderers-logo-480x480.png"
}

# ==========================================================
# 1. Imports
# ==========================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D

# ==========================================================
# 2. Combine All Data
# ==========================================================

def to_df(data, colname):
    return pd.DataFrame(data, columns=["Club", colname]).set_index("Club")

df = pd.concat([
    to_df(goals_scored_change, "Goals_Scored_Change"),
    to_df(goals_conceded_change, "Goals_Conceded_Change"),
    to_df(yellow_cards_change, "Yellow_Cards_Change"),
    to_df(xg_change, "xG_Change"),
    to_df(xg_against_change, "xG_Against_Change"),
    to_df(xg_difference_change, "xG_Diff_Change"),
    to_df(club_spend_24_25, "Spend"),
    to_df(income_24_25, "Income"),
    to_df(net_spend_24_25, "Net_Spend"),
    to_df(transfers_in_2425, "Transfers_In"),
    to_df(transfers_out_2425, "Transfers_Out"),
    to_df(xpoints_change, "xPoints_Change")
], axis=1)

# ==========================================================
# 3. Correlation Matrices
# ==========================================================

pearson_corr = df.corr(method="pearson")
spearman_corr = df.corr(method="spearman")

plt.figure(figsize=(12,10))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", center=0)
plt.title("Pearson Correlation Matrix")
plt.show()

plt.figure(figsize=(12,10))
sns.heatmap(spearman_corr, annot=True, cmap="viridis", center=0)
plt.title("Spearman Correlation Matrix")
plt.show()

# ==========================================================
# 4. Standardized OLS Sensitivity
# ==========================================================

X = df.drop(columns=["xPoints_Change"])
y = df["xPoints_Change"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = sm.add_constant(X_scaled)

model = sm.OLS(y, X_scaled).fit()
print(model.summary())

coefs = model.params[1:]

plt.figure(figsize=(10,6))
plt.barh(X.columns, coefs)
plt.axvline(0)
plt.title("Standardized Sensitivity vs xPoints Change")
plt.show()

# ==========================================================
# 5. Crest Plot Utility (Absolute Path + Safe)
# ==========================================================

def add_crests(ax, x_vals, y_vals):
    for club, x, y in zip(df.index, x_vals, y_vals):

        file_path = os.path.abspath(crest_paths[club])

        if not os.path.exists(file_path):
            print("Missing file:", file_path)
            continue

        img = plt.imread(file_path)

        zoom = 0.05
        if club == "Bournemouth":
            zoom /= 2.5

        imagebox = OffsetImage(img, zoom=zoom)
        ab = AnnotationBbox(imagebox, (x, y), frameon=False)
        ax.add_artist(ab)

# ==========================================================
# 6. xG Diff vs xPoints
# ==========================================================

fig, ax = plt.subplots(figsize=(8,6))
x = df["xG_Diff_Change"]
y = df["xPoints_Change"]

ax.scatter(x, y, alpha=0)
add_crests(ax, x, y)

ax.axhline(0)
ax.axvline(0)
ax.set_xlabel("xG Difference Change")
ax.set_ylabel("xPoints Change")
ax.set_title("xG Difference vs xPoints Change")

plt.show()

# ==========================================================
# 7. Net Spend vs xPoints
# ==========================================================

fig, ax = plt.subplots(figsize=(8,6))
x = df["Net_Spend"]
y = df["xPoints_Change"]

ax.scatter(x, y, alpha=0)
add_crests(ax, x, y)

ax.axhline(0)
ax.axvline(0)
ax.set_xlabel("Net Spend (m)")
ax.set_ylabel("xPoints Change")
ax.set_title("Net Spend vs xPoints Change")

plt.show()

# ==========================================================
# 8. Goals Scored vs xPoints
# ==========================================================

fig, ax = plt.subplots(figsize=(8,6))
x = df["Goals_Scored_Change"]
y = df["xPoints_Change"]

ax.scatter(x, y, alpha=0)
add_crests(ax, x, y)

ax.axhline(0)
ax.axvline(0)
ax.set_xlabel("Goals Scored Change")
ax.set_ylabel("xPoints Change")
ax.set_title("Goals Scored Change vs xPoints")

plt.show()

# ==========================================================
# 9. Ranked Correlation with xPoints
# ==========================================================

corr_with_xpoints = pearson_corr["xPoints_Change"].drop("xPoints_Change")
corr_with_xpoints = corr_with_xpoints.sort_values()

plt.figure(figsize=(10,6))
corr_with_xpoints.plot(kind="barh")
plt.axvline(0)
plt.title("Correlation Strength with xPoints Change")
plt.show()
```

```python
# ==========================================================
# 0. Crest Dictionary
# ==========================================================

crest_paths = {
    "Arsenal": "Arsenal-FC-logo-480x480.png",
    "Aston Villa": "Aston-Villa-FC-logo-480x480.png",
    "Bournemouth": "AFC-Bournemouth-logo-480x480.png",
    "Brentford": "Brentford-FC-logo-480x480.png",
    "Brighton": "Brighton-Hove-Albion-logo-480x480.png",
    "Chelsea": "Chelsea-FC-logo-480x480.png",
    "Cry. Palace": "Crystal-Palace-FC-logo-480x480.png",
    "Everton": "Everton-FC-logo-480x480.png",
    "Fulham": "Fulham-FC-logo-480x480.png",
    "Liverpool": "Liverpool-FC-logo-480x480.png",
    "Man. City": "Manchester-City-FC-logo-480x480.png",
    "Man. Utd": "Manchester-United-FC-logo-480x480.png",
    "Newcastle": "Newcastle-United-logo-480x480.png",
    "Nott Forest": "Nottingham-Forest-FC-logo-480x480.png",
    "Spurs": "Tottenham-Hotspur-logo-480x480.png",
    "West Ham": "West-Ham-United-FC-logo-480x480.png",
    "Wolves": "Wolverhampton-Wanderers-logo-480x480.png"
}

# ==========================================================
# 1. Imports
# ==========================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# ==========================================================
# 2. Combine Data
# ==========================================================

def to_df(data, colname):
    return pd.DataFrame(data, columns=["Club", colname]).set_index("Club")

df = pd.concat([
    to_df(goals_scored_change, "Goals_Scored_Change"),
    to_df(goals_conceded_change, "Goals_Conceded_Change"),
    to_df(yellow_cards_change, "Yellow_Cards_Change"),
    to_df(xg_change, "xG_Change"),
    to_df(xg_against_change, "xG_Against_Change"),
    to_df(xg_difference_change, "xG_Diff_Change"),
    to_df(club_spend_24_25, "Spend"),
    to_df(income_24_25, "Income"),
    to_df(net_spend_24_25, "Net_Spend"),
    to_df(transfers_in_2425, "Transfers_In"),
    to_df(transfers_out_2425, "Transfers_Out"),
    to_df(xpoints_change, "xPoints_Change")
], axis=1)

# ==========================================================
# 3. Crest Plot Utility
# ==========================================================

def add_crests(ax, x_vals, y_vals):
    for club, x, y in zip(df.index, x_vals, y_vals):

        file_path = os.path.abspath(crest_paths[club])
        if not os.path.exists(file_path):
            print("Missing file:", file_path)
            continue

        img = plt.imread(file_path)

        zoom = 0.08
        if club == "Bournemouth":
            zoom /= 2.5

        imagebox = OffsetImage(img, zoom=zoom)
        ab = AnnotationBbox(imagebox, (x, y), frameon=False)
        ax.add_artist(ab)

# ==========================================================
# 4. Correlation Plot Function
# ==========================================================

def correlation_plot(x_col, y_col, xlabel, ylabel, title):

    x = df[x_col].values
    y = df[y_col].values

    r = np.corrcoef(x, y)[0,1]

    slope, intercept = np.polyfit(x, y, 1)
    x_line = np.linspace(min(x), max(x), 200)
    y_line = slope * x_line + intercept

    fig, ax = plt.subplots(figsize=(12,9))

    ax.scatter(x, y, alpha=0)
    add_crests(ax, x, y)

    ax.plot(x_line, y_line, color="red",
            linewidth=2.5,
            label=f"Line of Best Fit (r = {r:.3f})")

    ax.grid(True)

    ax.set_xlabel(xlabel, fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.set_title(title, fontsize=16)

    ax.legend(fontsize=12)

    plt.tight_layout()
    plt.show()

# ==========================================================
# 5. Correlation Graphs
# ==========================================================

correlation_plot(
    "xG_Diff_Change",
    "xPoints_Change",
    "xG Difference Change",
    "xPoints Change",
    "xG Difference vs xPoints"
)

correlation_plot(
    "Net_Spend",
    "xPoints_Change",
    "Net Spend (m)",
    "xPoints Change",
    "Net Spend vs xPoints"
)

correlation_plot(
    "Goals_Scored_Change",
    "xPoints_Change",
    "Goals Scored Change",
    "xPoints Change",
    "Goals Scored Change vs xPoints"
)

correlation_plot(
    "Goals_Conceded_Change",
    "xPoints_Change",
    "Goals Conceded Change",
    "xPoints Change",
    "Goals Conceded Change vs xPoints"
)

correlation_plot(
    "Transfers_In",
    "xPoints_Change",
    "Transfers In",
    "xPoints Change",
    "Transfers In vs xPoints"
)

correlation_plot(
    "Transfers_Out",
    "xPoints_Change",
    "Transfers Out",
    "xPoints Change",
    "Transfers Out vs xPoints"
)
```
